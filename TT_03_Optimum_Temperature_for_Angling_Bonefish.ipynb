## 2. Purpose

The purpose of this notebook is to equip the local angling eco-tourism community with a pre-configured executable notebook capable of revealing scientific insight into their fishing grounds.

Novel approaches in flats ecology allow scientists to understand the bonefish's behavioral response to changes in sea surface temperature. Water temperatures play an extremely important role in bonefish (Albula spp.) movements and feeding patterns in coastal flats. Studies suggest that bonefish occupy habitats that approach their critical thermal maximum. The thermal maximum represents temperature limits lethal to bonefish. As benthivorous fish, bonefish migrate into shallow water habitats at low tide for food and migrate out of shallow water habitats at high tide. Due to their migration pattern bonefish are a species in sport-fishing. In the Bahamas, local communities depend on the influx of tourists looking to catch and release bonefish in the wetlands. The Marls is the largest and one of the most popular angling sites in the Bahamas. It contains over 300 square miles of wetlands and mangrove forests. Local eco-tourism businesses utilize weather conditions, tide activity, and seasonal temperature variations to determine the best areas to fish for bonefish. Earth climate models provide local businesses with this data. However, the optimal temperature is estimated from laboratory and field experiments. Advancements in cloud computing like Pangeo tools and data containers like dask provide the opportunity to combine these observations for making better decisions about the fishing grounds. In addition, access to scientific analysis tools also gives businesses insight into responsible fishing practices.

Large public datasets containing earth climate data provide users free access to apply their findings with minimal cost and effort. Analyzing these datasets can uncover global and regional patterns. This notebook will be utilizing Pangeo ecosystem tools to contain a large dataset for analysis. Loading this dataset with a container like xarray can provide users a variety of ways to visualize patterns and make future predictions. This notebook will analyze optimal temperatures for migrating bonefish within large datasets. This notebook uses the Multi-Ultra High-Resolution Seas Surface Temperature dataset from AWS as a case study to identify optimal temperature grids for bonefish migrating nearshore in the Marls, Abaco.

### 12.3. Examine Metadata in MUR Dataset

Zarr, a python library, is a unique data format for N-dimensional arrays that provides convenient access to large datasets. Zarr splits data into chunks, compresses each chunk into a zip file, and stores data with metadata into a directory. Zarr arrays work congruently with NumPy and similar libraries. Here, Xarray reads and stores metadata and variables from the Multi-Scale Ultra-High Resolution (MUR) Sea Surface Temperature (SST) dataset stored in the Zarr format. The variable metadata is useful for understanding the dataset's dimensions, coordinates, and data variables. This dataset has four data variables: analyzed_sst, analysis error, mask, and sea ice fraction.

```python
%%time
#load MUR dataset contained in zarr data format using xarray
ds = xr.open_zarr('https://mur-sst.s3.us-west-2.amazonaws.com/zarr-v1',consolidated=True)
#print ds variable to examine metadata
ds
```

```python
#set lattitude and longitude to the region around the Marls. 
lon_min = -78
lon_max = -77
lat_min = 26
lat_max = 27
my_time = '2010-01-01'

#subset MUR dataset in time and space around the Marls, Abaco for one date.
ds.analysed_sst.sel(time=my_time, method='nearest').sel(lon=slice(lon_min, lon_max),
                                                                 lat=slice(lat_min, lat_max)).plot() #how to use nearest in this situation
```

Notice the analyzed sea surface temperature is in Kelvin. With python numerical tools, we can convert Kelvin into Farenheit. With xarray, we can add the data array back into the dataset.

```python
#convert SST in Kelvin to Farenheit. 
temp_f = (ds.analysed_sst - 273.15) * 1.8 + 32
```

```python
#subset dataset in time and space. "squeeze()" to remove axis of length one. In this case, it is time.
marls_sst=temp_f.sel(time='2015-07-04', method='nearest').sel(lon=slice(lon_min, lon_max),
                                                                 lat=slice(lat_min, lat_max)).squeeze()
```

Now, let's look at the sea surface temperature with the temperature units as Fahrenheit. Matplotlib.pyplot.contourf plots filled contours. The number and region of contour lines are determined by level parameters.

```python
#contourf plot with the region of contour lines set to sst data stored in marls_sst variable.
plt.contourf(marls_sst.lon, marls_sst.lat, marls_sst, cmap='inferno')
plt.title('Date = ' + " " + my_time)
cbar = plt.colorbar()
plt.xlabel('longitude')
plt.ylabel('latitude')
cbar.set_label('temperature [Fahrenheit]')
```

```python
#plot sst
plt.pcolormesh(marls_sst.lon, marls_sst.lat, marls_sst)
#overlay plot with the number contour lines set to the optimal temperature range and the region defined by marls_sst.
plt.contourf(marls_sst.lon, marls_sst.lat, marls_sst, levels=optemp, cmap = 'PuRd')
cbar = plt.colorbar()
plt.xlabel('longitude')
plt.ylabel('latitude')
cbar.set_label('temperature [Fahrenheit]');
```

```python
#subset dataset in space to the entire Marls and surrounding region. No specfic date.
marls=temp_f.sel(lon=slice(lon_min, lon_max),lat=slice(lat_min, lat_max)).squeeze()

#visualize variability in optimal temperature range in the entire Marls region. 
marls.hvplot.contourf(levels=optemp, cmap = 'PuRd', clabel = '°[F]', geo=True, 
                                    widget_location='left_top', 
                                    tiles='CartoEco', alpha = 0.80, transform=ccrs.PlateCarree())
```

### 12.4. Marls Climatology (MUR)

If we want to look at the specific temperature variation at one location over the course of the year, we can create a "climatology" plot. These plots illustrate the average temperature for each month or each day. 

**Step 1.** Select a fishing site by indexing a single point in the data. Also, group the data by day. Xarray API provides users the ability to perform grouped operations. This is particularly useful to observe the changes in temperature at a single angling site throughout the year.

```python
#select range of years to examine climate variability
my_time="2002,2020"
#create variable to examine temperature variability at one site within the Marls. Temperature in Farenheit.
angling_site=temp_f.sel(lon=(-77),lat=(27), method = 'nearest')
#average temperature data by day
site_temp = angling_site.groupby("time.dayofyear").mean('time')
#examine metadata
site_temp
```

The plot below illustrates the monthly variations in temperature at one site. This is done by taking the average temperature of one day and plotting the temperature as a function of time. From this plot, it is evident that temperatures at this site are within the optimal temperature range. You can also see warmer sea surface temperatures during the summer months and a steady decline into the fall and winter months.   

```python
site_temp.plot()
plt.xlabel('Days of the Year')
plt.ylabel('SST degrees [Fahrenheit]');
```

**Step 2.** Calculate anomalies. A positive anomaly indicates the observed temperature is warmer than the average. A negative anomaly indicates the observed temperature is colder than the average. Using xarray group operations functionality it is possible to calculate the temperature variation from the average temperature. Xarray.groupby creates a new dataset with values grouped along a specified dimension. Xarray API also provides a function to apply vectorized functions to a grouped dataset. This function also compatible with dask to allow for more efficient processing. 

```python
#average temperature variability by day
daily_climatology_mean = angling_site.groupby("time.dayofyear").mean("time")
daily_climatology_std = angling_site.groupby("time.dayofyear").std("time")
# the distance between the temperature and the daily average
daily_stand_anomalies = xr.apply_ufunc(
    lambda x, m, s: (x - m) / s,
    angling_site.groupby("time.dayofyear"),
    daily_climatology_mean,
    daily_climatology_std,
    input_core_dims=None, dask = 'allowed', vectorize = True
)
```

```python
#plot
fig, (ax1, ax2) = plt.subplots(2, 1)

daily_climatology_mean.plot(ax=ax1)
plt.xlabel('Days of the Year')
plt.ylabel('SST degrees [Fahrenheit]')

daily_stand_anomalies.plot(ax=ax2)
plt.xlabel('Years')
plt.ylabel('Anomaly [Fahrenheit]')
plt.tight_layout();
```

```python
#average temperature variability by month
monthly_climatology_mean = angling_site.groupby("time.month").mean("time")
monthly_climatology_std = angling_site.groupby("time.month").std("time")
# the distance between the temperature and the monthly average
monthly_stand_anomalies = xr.apply_ufunc(
    lambda x, m, s: (x - m) / s,
    angling_site.groupby("time.month"),
    monthly_climatology_mean,
    monthly_climatology_std,
    input_core_dims=None, dask = 'allowed', vectorize = True
)
```

```python
#plot
fig, (ax1, ax2) = plt.subplots(2, 1)

monthly_climatology_mean.plot(ax=ax1)
plt.xlabel('Months')
plt.ylabel('SST degrees [Fahrenheit]')

monthly_stand_anomalies.plot(ax=ax2)
plt.xlabel('Years')
plt.ylabel('Anomaly [Fahrenheit]')
plt.tight_layout(pad=3.0);
```
